```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from scipy import sparse
```

```python
karate = nx.karate_club_graph() #Zachary's karate club
```

```python
nx.draw_circular(karate, with_labels=True)
```

```python
karate_laplacian = nx.linalg.laplacianmatrix.laplacian_matrix(karate)
```

```python
karate_laplacian
```

```python
plt.spy(karate_laplacian) 
#in addition to existing edges, since (l)_ij is just -a_ij, 
#this sparsity plot also has diagonal elements filled in (as (l)_ii is d_i)
plt.savefig('karatelaplacian.png')
```

```python
kl = sparse.csr_matrix.toarray(karate_laplacian)
eigenvalues = np.linalg.eig(kl)[0]
eigenvalues
```

```python
eigenvalues.sort()
eigenvalues #0 is simple -> strongly connected graph
```

```python
degree = 34
nodes = list(range(0, degree))
```

```python
fiedler = np.linalg.eig(kl)[1][10] #2nd eigenvalue
zipped_lists = zip(fiedler, nodes)
sorted_lists = sorted(zipped_lists)
sorted_nodes = list([element for _, element in sorted_lists])
fiedler = sorted(fiedler)
```

```python
fiedler
```

```python
def first_pos(l):
    #returns index of first positive element of sorted list of eigenvalues.
    #TODO: make binary
    for i in range(0,len(l)):
        if fiedler[i] > 0:
            print(i)
            break
        i+=1
```

```python
divider = first_pos(fiedler)
```
